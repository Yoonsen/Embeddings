```python
import folium
import json
```

```python
def make_map(places) :
    m = folium.Map(location=[20, 0], zoom_start=2)
    for place, info in places.items():
        if 'latitude' in info and 'longitude' in info:
            folium.Marker(
                location=[info['latitude'], info['longitude']],
                popup=f"<b>{place}</b><br>{info['description']}"
            ).add_to(m)
    return m
```

```python
make_map(places).save('kystby.html')
```
